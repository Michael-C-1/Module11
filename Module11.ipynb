#Combined Cycle Power Plant(CCPP) 6-year Data
This dataset predicts the net hourly electrical energy output(EP) based on 5 hourly average variables. The set has 9568data points collected from a CCPP over 6 years, when the power plant was set to work with full load. The variables are Temperature(T), Ambient Pressure(AP), Relative Humidity(RH) and Exhaust Vacuum(V). Each variable has its own range.
T is in the range 1.81-37.11 degrees Celcius. AP is in the range 992.89-1033.30millibars. RH is in the range 25.66-100.16%. V is in the range 25.36-81.56 cm Hg. EP is in the range 420.26-495.76MW.

```
import numpy as nm
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/Michael-C-1/Module10/main/Module10%20data%20set.csv')
df.head()
```

##Check for nulls
This checks to see if there are missing values in the dataset.

```
df.isnull().sum()
```

###Removing null rows
This will remove missing vales from the dataset, if there are any.

```
df = df.dropna(how='any',axis=0)
df.isnull().sum()
```

##Describe
This is a deescription of the set, which includes the mean value, standard deviation, minimum value, quartiles, and maximum value for each variable.

```
df.describe()
```

##Graphing data
This makes it so that the graphs of the variables can be compared

```
df.iloc[:,:5].plot.kde()
for col in df:
  sns.kdeplot(df[col], shade=True)
```

###Normalized graph
This makes it so that for each variable, the mean value is subtracted from a data point and the result is divided by the standard deviation. This makes it so that the mean for all of the variables is 0 and they all have a similar scale.

```
normalized_df=(df-df.mean())/df.std()
normalized_df.iloc[:,:5].plot.kde()
for col in df:
  sns.kdeplot(normalized_df[col], shade=True)
```

##Histograms
This shows what distribution the variables have.

```
df.hist(bins=50, figsize=(15,10))
plt.show()
```

V, T, and EP seem to have bimodal distributions. RH has a skewed distribution. AP looks like it has a normal distribution.

##Correlation Matrix
This tells us how the variables are related. The further from 0, the higher the correlation.

```
corr = df.corr()
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(data=corr, square=True , annot=True, cbar=True, ax = ax)
```

T and V seem to have a high positive correlation and both have an extremely high negative correlation to EP. This means that as T increases, V increases and EP decreases

##Summary
In conclusion, there are 5 variables in the dataset. Of these, 2 have an extremely high negative correlation with EP: T and V. T and V also have a high positive correlation with each other. Coincidentally, those three are also the variables with bimodal distributions.

#Wine Data Set
This set has over 12,700 wines, with many of the attributes being related to the chemical composition of the wines. The attributes are: TARGET, AcidIndex, Alcohol, Chlorides, Density, FixedAcidity, LabelAppeal, ResidualSugar, STARS, Sulphates, TotalSulfurDioxide and VolatileAcidity. TARGET is the # of cases of wine sold. AcidIndex measures the total acidity of wine via a weighted average. Alcohol is the alcohol content.

Chlorides is the chloride content of the wine. Density is the density of the wine. FixedAcidity is the FixedAcidity of the wine. LabelAppeal is the subjective marketting score that indicates the appeal of the design of the label on the bottle. ResidualSugar is the residual sugar content of the wine.

STARS is the wine rating as determined by experts (4=excellent, 1=poor). Sulphates is the Sulfate content of the wine. TotalSulfurDioxide is the total sulfur dioxide content of the wine. VolatileAcidity is the volatile acid content of the wine. pH is the pH of the wine.
